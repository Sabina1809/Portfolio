```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = [[1], [2], [5], [6]]
y = ["cat", "cat", "dog", "dog"]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

print(model.predict([[3]]))
```
